```python
import cv2
import numpy as np
from PIL import Image ,ImageEnhance
from PIL import ImageFilter
import matplotlib.pyplot as plt
import pytesseract
```

```python
img = Image.open("Epo Automation Content.jpeg")

#enhance edge
edge_enhanced = img.filter(ImageFilter.EDGE_ENHANCE)

plt.imshow(edge_enhanced)
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original")
plt.subplot(1,2,2)
plt.imshow(edge_enhanced)
plt.title("Enhanced")
edge_enhanced.save("enhanced.png")
```

```python
#threshold to binary inverstion
img= Image.open("enhanced.png") #read image
gray = img.convert('L')
bw = gray.point(lambda x: 0 if x<128 else 255, '1')
bw.save("black_white.png")

#enhance edge
img= Image.open("black_white.png") #read image
edge_enhanced = img.filter(ImageFilter.EDGE_ENHANCE)
edge_enhanced.save("final_enhanced.png")


plt.subplot(1,2,1)
plt.imshow(bw)
plt.title("Black_White")
plt.subplot(1,2,2)
plt.imshow(edge_enhanced)
plt.title("Enhanced ")

```

```python
#passing image to ocr
string=pytesseract.image_to_string(Image.open("enhanced.png"))
string.encode("ascii","replace")
```
