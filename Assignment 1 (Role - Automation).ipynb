```python
import pandas as pd
import numpy as np
from datetime import datetime
dataframe=pd.read_csv("Automatic email content store.csv",delimiter="$")
dataframe.head() #topp 5 rows
```

```python
#function to find unique email address
def email_address_list(dataframe):
    email_dictionary={}
    for itr in range (0,len(dataframe)):
        if dataframe["to "][itr] in email_dictionary:
            email_dictionary[dataframe["to "][itr]]+=1
        else :
            email_dictionary[dataframe["to "][itr]]=1
    email_list=[]
    for key in email_dictionary.iterkeys():
        email_list.append(key)
    return email_list

print "Unique Email Address: \n", np.asarray(email_address_list(dataframe)).reshape(-1,1)
```

```python
#function to find the number of delays elapsed
def days_elapsed(dataframe):
    dataframe["days_elapsed"]=""
    for itr3 in range(0,len(dataframe)):    
        current_date=datetime(datetime.now().year,datetime.now().month,datetime.now().day)
        try:
            past_date=dataframe["Date"][itr3].split("/")
            past_date=datetime((2000+int(past_date[2])),int(past_date[1]),int(past_date[0]))
            dataframe["days_elapsed"][itr3]=(current_date-past_date).days
        except:
            pass
        
days_elapsed(dataframe)
dataframe.head()     
```

```python
#function to find word
def find_word_content(dataframe):
    for itr4 in range(0,len(dataframe)):
        content=str(dataframe["raw"][itr4]).split(" ")
        keyword_list=["Python","OpenCv","Machine Learning"]
        dataframe["Keyword_present"]=""
        for word in content:
            if word in keyword_list:
                dataframe["Keyword_present"][itr4]= dataframe["Keyword_present"][itr4] + " " + word

find_word_content(dataframe)
dataframe.head()
```
